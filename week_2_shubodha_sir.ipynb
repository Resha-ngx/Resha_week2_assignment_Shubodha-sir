```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/spotify_dataset.csv')
```

```python
# Basic Information
print("First 5 rows of the dataset:")
print(df.head())
```

```python
print("\nSummary of the dataset:")
print(df.info())
```

```python
print("\nStatistical summary of numerical columns:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Data Types and Column Names
print("\nColumn names and data types:")
print(df.dtypes)
```

```python
# Selecting 'Artist', 'Song Name', 'Genre', and 'Popularity' columns
selected_columns = df[['Artist', 'Song Name', 'Genre', 'Popularity']]

# Display the selected columns
print(selected_columns.head())
```

```python

# Convert 'Popularity' to numeric, handling errors
df['Popularity'] = pd.to_numeric(df['Popularity'], errors='coerce')

# Plotting histogram for the 'Popularity' column
df['Popularity'].dropna().hist(bins=15, figsize=(8, 5))
plt.title("Histogram of Popularity")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()


```

```python
# Count plots for categorical columns: 'Artist', 'Song Name', and 'Genre'
categorical_columns = ['Artist', 'Song Name', 'Genre']
for col in categorical_columns:
    plt.figure(figsize=(10, 5))
    top_categories = df[col].value_counts().nlargest(10)  # Show top 10 most frequent categories for readability
    sns.barplot(x=top_categories.index, y=top_categories.values, palette="viridis")
    plt.title(f"Top 10 {col} by Frequency")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Selecting multiple numerical columns
selected_columns = ['Highest Charting Position', 'Number of Times Charted', 'Popularity']

# Convert columns to numeric where needed, handling non-numeric values with coercion
for col in selected_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calculate and plot the correlation matrix for the selected numerical columns
plt.figure(figsize=(8, 6))
corr_matrix = df[selected_columns].corr()  # Calculate correlation matrix for selected columns only
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix for Selected Numerical Columns")
plt.show()
```

```python
# Pair plot for 'Week of Highest Charting', 'Highest Charting Position', 'Number of Times Charted', and 'Popularity'
selected_columns = [ 'Highest Charting Position', 'Number of Times Charted', 'Popularity']

# Convert the columns to numeric to ensure they can be plotted correctly
for col in selected_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Creating the pair plot
pair_plot = sns.pairplot(df[selected_columns])
pair_plot.fig.suptitle("Pair Plot of Selected Numerical Features", y=1.02)

plt.show()
```

```python


# Specify the numerical columns you want to visualize
selected_numerical_columns = ['Highest Charting Position', 'Number of Times Charted', 'Popularity']

# Outlier detection using boxplots for the selected numerical columns
for col in selected_numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, x=col)
    plt.title(f"Box plot of {col}")
    plt.show()



```
